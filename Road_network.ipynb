```python
import matplotlib.pyplot as plt
import osmnx as ox
from shapely.geometry import box
import geopandas as gpd

# Define the bounding box coordinates for Agartala
north, south, east, west = 23.88, 23.77, 91.35, 91.22

# Create a rectangular polygon from the bounding box coordinates
admin_poly = box(west, south, east, north)

# Create a GeoDataFrame from the polygon
admin = gpd.GeoDataFrame(index=[0], crs="EPSG:4326", geometry=[admin_poly])

# Extract the graph for the area within the rectangle
G = ox.graph_from_polygon(admin_poly)

# Convert graph to GeoDataFrames
edges = ox.graph_to_gdfs(G, nodes=False)


```

```python
# Your existing plotting code
f, ax = plt.subplots(1, figsize=(10, 10), facecolor='black')
edges.plot(ax=ax, edgecolor='red', linewidth=0.45)

# Add title to the map
ax.set_title('Street Map of Agartala', fontsize=16, color='white')

# Hide axis
ax.axis('off')

# Save the figure
plt.savefig('Map.jpeg', dpi=900)

# Show the figure
plt.show()
```
